```python
c.rewind()  # just in case
et = pytz.timezone('US/Eastern')
deltaoneday = datetime.timedelta(1)

#this is the wd (0=Monday, Sunday=6) corresponding to the earliest it *could*
#have been submitted (e.g., if submitted 11am on Tuesday, this is Monday=0,
#but if 4:01pm, it's Tuesday=1
earliestsubwd = []
ncites = []

for i,d in enumerate(c):
    dtutc = d['arxiv_date'].replace(tzinfo=pytz.UTC)
    dtet = dtutc.astimezone(et)
    if dtet.hour <16:
        date = (dtet - deltaoneday).date()
    else:
        date = dtet.date()
    earliestsubwd.append(date.weekday())
    ncites.append(d['ncites'])
earliestsubwd = np.array(earliestsubwd)
ncites = np.array(ncites)
c.rewind();
```

```python
# this is the wd (0=monday, 4=friday) when it actually appears in the morning 
# mailing.  This -1 at 8pm is when it appears on the web site
wdlisted = earliestsubwd + 2
wdlisted[wdlisted==5] = 0 # submitting on Th -> Monday
wdlisted[wdlisted>5] = 1 # submitting Fri-Sun -> Tues
```

```python
def plotcdf(n, **kwargs):
    cnts = np.bincount(n)
    cnts = np.concatenate((cnts,[0]))
    x = np.arange(len(cnts))
    x[-1] = x[-2]
    return plt.plot(x, cnts, linestyle='steps-mid', **kwargs)
```

```python
figure(figsize=(14,10))
plotcdf(ncites[earliestsubwd==0],label='Mon');
plotcdf(ncites[earliestsubwd==1],label='Tu');
plotcdf(ncites[earliestsubwd==2],label='Wed');
plotcdf(ncites[earliestsubwd==3],label='Th');
plotcdf(ncites[earliestsubwd==4],label='Fri');
plotcdf(ncites[earliestsubwd==5],label='Sat');
plotcdf(ncites[earliestsubwd==6],label='Sun');
loglog();
legend();
title('When submitted (after 4pm on)')
xlabel('citations')
ylabel('N');
```
